## Variable Description

page_id: The unique identifier for that characters page within the wikia

name: The name of the character

urlslug: The unique url within the wikia that takes you to the character

ID: The identity status of the character (Secret Identity, Public identity)

ALIGN: If the character is Good, Bad or Neutral

EYE: Eye color of the character

HAIR: Hair color of the character

SEX: Sex of the character (e.g. Male, Female, etc.)

GSM: If the character is a gender or sexual minority (e.g. Homosexual characters, bisexual characters)

ALIVE: If the character is alive or deceased

APPEARANCES: The number of appareances of the character in comic books (as of Sep. 2, 2014. Number will become increasingly out of date as time goes on.)

FIRST APPEARANCE: The month and year of the character's first appearance in a comic book, if available
YEAR: The year of the character's first appearance in a comic book, if available

## Data Importing and Reading

```python
dc_filepath = "dataset/dc-wikia-data.csv"
dc_df = pd.read_csv(dc_filepath)

dc_df.head(10)
```

## Inspection of null values in each column

```python
null_vals = dc_df.isnull().sum()
null_vals_df = pd.DataFrame({'column':null_vals.index, 'count':null_vals.values})
null_vals_df['percentage'] = (null_vals_df['count']/dc_df.shape[0])*100
null_vals_df
```

## Subseting and Renaming the columns

```python
dc_df.columns
```

```python
dc_df = dc_df[['page_id', 
               'name',  
               'ID', 
               'ALIGN', 
               'EYE', 
               'HAIR', 
               'SEX',
               'ALIVE', 
               'APPEARANCES', 
               'YEAR']].copy()

columns_renamed = {'ID':'identity_status', 
               'ALIGN':'align', 
               'EYE':'eye_color', 
               'HAIR':'hair_color', 
               'SEX':'gender', 
               'ALIVE':'alive', 
               'APPEARANCES':'appearences', 
               'YEAR':'year_introduced'}

dc_df = dc_df.rename(columns=columns_renamed)
```

## Filling and droping missing values

```python
dc_df['identity_status'] = dc_df['identity_status'].fillna('Missing Identity')
dc_df['align'] = dc_df['align'].fillna('Side Characters')
dc_df['eye_color'] = dc_df['eye_color'].fillna('Other')
dc_df['hair_color'] = dc_df['hair_color'].fillna('Other')
dc_df['gender'] = dc_df['gender'].fillna('Other')
dc_df['alive'] = dc_df['alive'].fillna('Missing')
```

```python
# Imputing missing values in appearences column with mean of all the appearences
mean_appearence = dc_df['appearences'].mean()
dc_df['appearences']=dc_df['appearences'].fillna(mean_appearence)
```

```python
# Droping rows with missing vlaues that cannot be imputed
dc_df= dc_df.dropna(subset=['year_introduced'])
```

## Data Visualization

```python
# Importing required libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
sns.set_style('darkgrid')
```

### Year Introduced 

```python
ax = dc_df['year_introduced'].value_counts().head(20)\
    .plot(kind='barh', 
          title='Top 20 Years DC Characters Introduced in')

ax.set_ylabel('Year Introduced')
ax.set_xlabel('Number of Characters')
plt.show()
```

* According to above figure, most of the characters in the dataset were introduced in 2006.

### Character Alignment

```python
year_align = dc_df[['year_introduced','align']]
year_align = year_align\
            .groupby(['year_introduced','align'])\
            ['year_introduced']\
            .agg('count')\
            #.tail(20)
    
year_align
```

```python
known = dc_df.query('identity_status != "Identity Unknown" and align !="Reformed Criminals"')
ax = sns.countplot(known,x='identity_status',hue='align',saturation=1)
ax.set_xlabel('Identity Status')
ax.set_ylabel('Number of Character')
plt.show()
```

```python
good = dc_df.query('align == "Good Characters" and alive !="Missing"')
fg = sns.FacetGrid(good, col='alive')
ax = fg.map_dataframe(sns.kdeplot, x='appearences',color='red')
fg.set_axis_labels('Appearences')
plt.show()
```

```python
ax = sns.catplot(good,x='alive',y='appearences')
ax.set_axis_labels('Alive', 'Appearences')
plt.show()
```

### Physical Traits

```python
eye_color_counts = dc_df['eye_color'].value_counts()
ax = eye_color_counts\
    .sort_values(ascending=True)\
    .plot(kind='barh', 
          title='Number of Characters by Eye Color')

ax.set_ylabel('Eye Color')
ax.set_xlabel('Number of Characters')
plt.show()
```

```python
print(eye_color_counts)
```

```python
low_counts = eye_color_counts.loc[lambda x: x<30]
ax = low_counts\
    .sort_values(ascending=True)\
    .plot(kind='bar', 
          title='Least Number of Characters by Eye Color',)

ax.set_xlabel('Eye Color')
ax.set_ylabel('Number of Characters')
plt.show()

print(low_counts)
```

```python
year_appearence_corr = dc_df[['year_introduced','appearences']].corr()
sns.heatmap(year_appearence_corr,annot=True)
plt.show()
```

```python
gender_count = dc_df['gender'].value_counts()

# Ploting the chart to clearly represent genders with corresponding counts in the legend
labels = gender_count.index.values
vals = gender_count.values
patches, texts = plt.pie(gender_count)
labels_vals_map = ['{0} - {1:n}'.format(i,j) for i,j in zip(labels, vals)]

sort_legend = True
if sort_legend:
    patches, labels_vals_map, dummy =  zip(*sorted(zip(patches, labels_vals_map, vals),
                                          key=lambda x: x[2],
                                          reverse=True))
plt.legend(patches, labels_vals_map, loc='center right', bbox_to_anchor=(-0.1, 1.),
           fontsize=8)
plt.title('Various Genders of DC Characters')
plt.show()
```
